```python
df['Model Name'] = df['Product Name'].str[6:15]

df.head()
```

```python
df.sort_values(by='Star Rating',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# distribution of star ratings
sns.histplot(df['Star Rating'], bins=10, kde=True)
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#average sale price for each brand

avg_price_by_brand = df.groupby('Brand')['Sale Price'].mean().sort_values(ascending=False)
print(avg_price_by_brand)
```

```python
# Get the top 5 products with the highest number of reviews

top_reviews = df.nlargest(5, 'Number Of Reviews')[['Product Name', 'Number Of Reviews']]
print(top_reviews)
```

```python
# Get the top 5 products with the highest discount percentage

top_discounted_products = df.nlargest(5, 'Discount Percentage')[['Product Name', 'Discount Percentage']]
print(top_discounted_products)
```
